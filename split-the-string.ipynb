```python
#分割在分割做法
str="""Amy is 5 years old,and her sister Mary is 2 years old.Ruth and Peter,their parents,have 3 kids""" #split every word inside the string use the split function
str_split=str.split(',')
str_split=''.join(str_split) #change list to string
str_split=str_split.split('.')
str_split #split後字串會變list,可以用''.join(str_split)轉回成字串
#但這邊很明顯,more and more string 黏在一起,這樣無法計算說有幾個字,分割在分割方法證明無用
```

```python
#解題想法,將. ,換成空白在分割
import string #需要使用string.punctuation
str="""Amy is 5 years old,and her sister Mary is 2 years old.Ruth and Peter,their parents,have 3 kids""" #split every word inside the string use the split function
for word in str:
    if(word in string.punctuation):
        str=str.replace(word,' ') #replace(old_string,new_string,max)
print(str) #檢查換成space的str,可以看到之前的comma的位置都換成space了
#可以將字串切割然後計算字數了
str=str.split()
sum=0
for word in str:
    if(word.isalpha()==True):
            sum+=1
print(sum)
```
