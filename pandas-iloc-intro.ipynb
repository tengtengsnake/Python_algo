先建立一個資料集

```python
mydict = [
{'a': 1, 'b': 2, 'c': 3, 'd': 4},
{'a': 100, 'b': 200, 'c': 300, 'd': 400},
{'a': 1000, 'b': 2000, 'c': 3000, 'd': 4000 }
]
```

```python
#轉為df資料
import pandas as pd
df=pd.DataFrame(mydict)
df
```

```python
df.iloc[0] #拿第 0 列
```

```python
# 0 and 1 row
df.iloc[[0,1]]
```

```python
df.iloc[:3] #從0取到2 只有一個就是取row
```

```python
df.iloc[[True,False,True]]## 用array 布林是否來決定那個row要，這裡是0,2 row true, 就拿這兩個row
```

```python
#也可以用lamda來決定。這裡決定要拿取資料集（x）的index可以被2整除的row
df.iloc[lambda x:x.index%2==0]
```

```python
# 除了row（列）, 也可以指定 col（行）
df.iloc[0,1] # 0 row 1 col   ,號分隔row 或是 col
```

```python
#0 and 2 row ,1 and 3 col
df.iloc[[0,2],[1,3]] #,號隔開row and col
```

```python
# 1 ~3 row 0~3 col
df.iloc[1:3,0:3]
```

```python
# all row with 0 and 1 col
df.iloc[:,[True,False,True,False]]
```

```python
#all row with 0 and 2 col
df.iloc[:,lambda df:[0,2]]
```
