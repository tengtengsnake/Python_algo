```python
class Account():
    def __init__(self,number,name):
        self.number=number
        self.name=name
        self.balance=0
    def deposit(self,amount): #存錢
        assert amount > 0,'必須是>0的integer'
        self.balance+=amount
    def withdraw(self,amount): #領錢
        assert amount > 0,'必須是>0的integer'
        if amount <=self.balance:
            self.balance-=amount
        else:
            raise RuntimeError('balance not enough')
a=Account('E122','Justin')
a.deposit(-1)
```
