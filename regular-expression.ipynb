### re.search() 用法re.search(pattern,string)
pattern 資料形式
string:要比對的文字

```python
#範例1 - 只保留日期：
import re
text1='今天的日期是：2020-07-01 14:02:48'
time=re.search(r'\d+-\d+-\d+ \d+:\d+:\d+',text1).group(0)
time
```

```python
#尋找特定規律
#只從文本的最開頭開始比對 match
import re

pattern=re.compile(r'<HTML>')
pattern.match("<HTML><head>")
re.match(r'<HTML>',"<HTML><head>")         
```

```python
#尋找特定規律
#從任意位置開始比對
pattern=re.compile(r'<HTML>')
pattern.search("a<HTML><head>")
```

```python
#提取符合規律的文字
#回傳符合規律的文字列表
pattern=re.compile(r'<HTML>') #pattern文本 定義規則
pattern.findall("<HTML>_123_<HTML>") #findall will return a list extract some string
```

```python
#finditer 的使用,跟findall很像,不過是return 一個callable _iterator
import re
pattern=re.compile(r'<HTML>')
results=pattern.finditer("<HTML>_123_<HTML>")
type(results) #回傳一個迭代器
for i in results:
    print(i)
```

```python
#修改文本
#根據條件分割文本 
pattern=re.compile(r"\W") #只要不是word
results=pattern.split("Beautiful is better than-upgly!")
results
```

```python
simple_string="""Amy is 5 years old,and her sister Mary is 2 years old.Ruth and Peter,their parents,have 3 kids"""
pattern=re.compile(r"\W")#\w any 字母與數字
results=pattern.split(simple_string)
print(results)
sum=0
for word in results:
    if(word.isalpha()==True):
        sum+=1
print(sum)
```

```python
#正則表達式-group
#group 與matchobject
m=re.match(r"(\w+)\s(\w+)","Isaac Newton, physicist")
print(m.group(0)) #the entire match
print(m.group(1,2))
```

```python
match=re.search(r'(hello) \1 \S+','This is a hello hello world')
print(match.group(0)) 
```

```python
#name group 幫group 命名,增加可讀性 語法: ?P<>
match=re.search(r"(?P<foo>hello \S+)",'This a a hello world!')
print(match.group('foo'))
```
