Python 海象操作符號(:=) Python 3.8

以下是我的簡單的理解：

本來是無法在Expression 裡面進行assignment的
如下：

```python
def close_app_returned_code():
    return -1
if(close_code=close_app_returned_code()==-1):
    print('app is closed')
#will cause error  
```

```python
def close_app_returned_code():
    return -1
if(close_code:=close_app_returned_code()==-1):
    print('app is closed')
```
